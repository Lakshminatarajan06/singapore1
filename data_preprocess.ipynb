```python
def plot(df, column):

    plt.figure(figsize=(20,5))
    plt.subplot(1,3,1)
    sns.boxplot(data=df, x=column)
    plt.title(f'Box Plot for {column}')

    plt.subplot(1,3,2)
    # df[column].plot(kind='hist')
    sns.histplot(data=df, x=column, kde=True)
    plt.title(f'Distribution PLot for {column}')

    plt.subplot(1,3,3)
    sns.violinplot(data=df, x=column)
    plt.title(f'Violin Plot for {column}')
    plt.show()
```

```python
# Treating the outliers using IQR

def outlier(df, column):
    percentile25=df[column].quantile(0.25)
    percentile75=df[column].quantile(0.75)
    iqr=percentile75-percentile25

    upper_limit=percentile75+1.5*iqr
    lower_limit=percentile25-1.5*iqr

    # Replace outliers with upper or lower limit
    df[column] = df[column].apply(lambda x: upper_limit if x > upper_limit else (lower_limit if x < lower_limit else x))
    return df[column]
```

```python
outlier(skew_df, 'storey_range')
outlier(skew_df, 'floor_area_sqm')
outlier(skew_df, 'resale_price')
outlier(skew_df, 'remaining_lease')
outlier(skew_df, 'Flat_old')
```

```python
# Checking the list of algorithm to have better prediction.

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
def best_algorithm(x_train, x_test, y_train, y_test, algorithm_class):

    # Instantiate the model
    model = algorithm_class()
    
    # Model building
    model.fit(x_train, y_train)

    y_train_pred = model.predict(x_train)
    y_test_pred = model.predict(x_test)

    training_accuracy = r2_score(y_train, y_train_pred)
    testing_accuracy = r2_score(y_test, y_test_pred)

    accuracy_metrics = {
        'algorithm': algorithm_class.__name__,
        'R2_Train': training_accuracy,
        'R2_Test': testing_accuracy
    }
    
    return accuracy_metrics

```

```python
result_1 = best_algorithm(x_train, x_test, y_train, y_test, LinearRegression)
print(result_1)

result_2 = best_algorithm(x_train, x_test, y_train, y_test, DecisionTreeRegressor)
print(result_2)

```

```python
result_3 = best_algorithm(x_train, x_test, y_train, y_test, RandomForestRegressor)
print(result_3)
```

```python
# Random Forest - gives more accuracy.

from sklearn import metrics
import numpy as np

model=DecisionTreeRegressor().fit(x_train, y_train)

y_pred=model.predict(x_test)

# metrics
print("Mean Absolute Error:", metrics.mean_absolute_error(y_test, y_pred))
print("Mean Squared Error:", metrics.mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error:", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))

# R2 Score
R2_Test = r2_score(y_test, y_pred)
print("R2 Score:", R2_Test)

```
